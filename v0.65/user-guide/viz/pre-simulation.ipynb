# Configuration Visualizations

Some PassengerSim visualization tools can operate on a Config object
directly, because they don't rely on the outputs of a simulation.
Some examples are shown here.

```python
import passengersim as pax

cfg = pax.Config.from_yaml(pax.demo_network("3MKT/DEMO"))
```

## Route Map

You can generate a map of all the legs in the network, or all the legs offered by
a particular carrier (here the maps look the same, as the 3MKT network is symmetric).

```python
cfg.fig_route_map()
```

```python
cfg.fig_route_map("AL1")
```

## Booking Curves

You can review the booking curves, which control the average fraction of each passenger type
arriving in each timeframe.

```python
cfg.fig_booking_curves()
```
